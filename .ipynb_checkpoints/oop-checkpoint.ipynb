## Encapsulation (Data Hiding)

Think of encapsulation like a capsule or a box that keeps everything together.
It means bundling data (variables) and methods (functions) that operate on the data into a single unit (object).
It hides the details of an object and only exposes what is necessary.

🛠 Example:

A car has an engine, but you don’t need to know how the engine works to drive the car. You just use the accelerator and brake.

```python
class Car:
    def __init__(self, brand):
        self.__brand = brand  # Private variable (hidden from outside)

    def get_brand(self):  # Public method to access brand
        return self.__brand

car = Car("Tesla")
print(car.get_brand())
# print(car.__brand)
```

## Abstraction (Hiding Complexity)

Abstraction is like a TV remote—you press buttons without knowing how the circuit works.
It hides complex logic and only exposes necessary parts to the user.
It makes the program simpler and easier to use.

🛠 Example:

Imagine a Payment class that allows users to make payments without knowing the internal bank transaction details.

```python
from abc import ABC, abstractmethod

class Payment(ABC):  # Abstract class
    @abstractmethod
    def pay(self, amount):
        pass  # Force subclasses to define this method

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} with Credit Card.")

payment = CreditCardPayment()
payment.pay(100)
```

## Inheritance (Reusability)

Inheritance allows one class (child) to use properties and methods of another class (parent).
It helps reduce duplicate code and enables code reuse.

🛠 Example:
A Dog and Cat class can inherit common properties from an Animal class.

```python
class Animal:
    def speak(self):
        return "Animal makes a sound"

class Dog(Animal):  # Dog inherits from Animal
    def speak(self):
        return "Bark!"

dog = Dog()
print(dog.speak())
```

## Polymorphism (One Name, Many Forms)

Polymorphism allows different objects to use the same method in different ways.
It enables flexibility in the code.

🛠 Example:
A Dog and Cat both have a speak() method, but they return different sounds.

```python
class Cat(Animal):
    def speak(self):
        return "Meow!"

animals = [Dog(), Cat()]
for animal in animals:
    print(animal.speak())
```

### OOP Summary: Instances, Attributes, Methods

1. Instance Attributes vs Instance Methods

+ Instance Attributes → Variables that belong to an object.

+ Instance Methods → Functions that belong to an object and usually work with its attributes.

```python
class Student:
    def __init__(self, name, age):   # Constructor
        self.name = name             # instance attribute
        self.age = age               # instance attribute

    def introduce(self):             # instance method
        return f"My name is {self.name}, and I am {self.age} years old."
```

2. Define and Call an Instance Method

- Defined like a normal function, but inside a class.

- Always has self as the first parameter.

- Called using the object (instance).

```python
s1 = Student("Faith", 25)   # create object
print(s1.introduce())       # calling instance method
```

```python
s2 = Student("James", 30)
print(s2.name)   # James
print(s2.age)    # 30
```

s1 and s2 have different attributes even though they come from the same class.

4. The self Variable

- self represents the current instance (object) of the class.

- It allows each object to keep its own data.

```python
class Car:
    def __init__(self, brand):
        self.brand = brand   # 'self' makes it unique per object
    
    def show_brand(self):
        return f"This car is a {self.brand}."
```

c1 = Car("Toyota")
c2 = Car("BMW")

print(c1.show_brand())  # This car is a Toyota.
print(c2.show_brand())  # This car is a BMW.


Without self, Python wouldn’t know which object’s attribute/method you’re referring to

5. Create an Instance of a Class

Creating an object = instantiating a class.

```python
student1 = Student("Faith", 25)   # instance of Student
```
